```python
import this

# 1. Value
# Value is a group of string, integer, float, boolean or all combined.

#=================================================================================================================

# String (str) (consist of letter(s))
# A value should be enclosed with either quotation marks "" or apostrophe '' (quotation mark is much more common)
# e.g. "Car" or 'Table'

# Let's try print the word "Car"
# You will need print command by typing print()
# e.g.

print("car")
print("Table")

# Both command will print each word
#=================================================================================================================

# Integer (int)
# Unlike String that consist of letters, Integer consist of numbers e.g. 1, 2 , 3, etc.
# However, String is just a solid numbers
# e.g.

print(1)
print(125)
# Notice, We didn't put any quotation mark and it still coming out without syntax error
# It's because Integer doesn't need any quotation mark
# Scriptwriter app or IDE will automatically detect a number as a string if we put NO quotation mark
#=================================================================================================================

# To check if it is really an Integer. We need a special command called $type
# e.g.

print(type(1))
print(type(125))
# You will see <class 'int'> it means that it is indeed an integer
# Now let's do an experiment to prove that anything with quotation mark is called string
# e.g.

print(type("125"))
# Now after running the code you'll see <class 'str'>. it is indeed not an Integer
#=================================================================================================================

# Float
# Almost similar to an Integer which include number. Float is a number decimal in it (12.5, 21.7, etc.)
# e.g.

print(12.5)
print(type(12.5))
# From command type we can see that 12.5 is float because it have decimal in it
#=================================================================================================================

# Boolean (bool)
# This is a binary value. That means boolean only have two values, TRUE or FALSE
# Usually boolean will come very handy whenever whe use the logic formula of IF and any variation of IFS
# e.g.

# We will try to make variables (It will be discuss in the second part).
# Variable 1 will be called $Car and Variable 2 will be called $Not_Car
# In coding we cannot use space if we want to make variable
# If we write $Not Car then the intepreter would detect two words as separate value
# So to make it easier, we will put underscore (e.g. Not_Car)

Car = True
Not_Car = False

if Car:
    print("This item is a Car")
else:
    print("This item is NOT a Car")
# To make it more useful, we will integrate all of it into a print statement with f-string and variable
#=================================================================================================================

# NOTES
# If you try to erase the quotation mark, you will get a syntax error.
# VS Code is smart enough to know what's wrong with your code
```

```python
# 2. Variable, F-String, and Input

#=================================================================================================================

# Variable
# This is a container of value
# A Variable will behave as it was the value
# e.g.

nationality = "Indonesia"

print(nationality)
# If we run the code it will print "Indonesia" even when we put nationality in the print statement
# It is because our variable $nationality is containing the value of "Indonesia"
# This function is useful to shortening a value like numbers or long words
# e.g.

# Let says we have an income as data, instead writing all over again in coding
# We just simply named the numbers as $Income. So we didn't have to look all over again for the value

income = 12120000
print(income)

#=================================================================================================================

```

```python
# 3. IFS / ELSE and Conditional Operators

#=================================================================================================================

# if and else
# The way this is works is to giving result with a certain condition
# For example: when you ride a rollercoaster it need a certain condition, which is height
# If the height > 140 cm then you can ride, else you couldn't
# The formula looks like this

# if condition:
#     do this
# else:
#     do this
# e.g. you need to reach > 140 cm to ride a rollercoaster

print("Welcome to Wonderia Theme Park!")
height = int(input("What is your height?"))

if height > 140:
    print("Congrats! You are able to ride the Wonder Rollercoaster!")
else:
    print("Sorry, you cannot ride the Wonder Rollercoaster!")
```
